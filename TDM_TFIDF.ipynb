```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
import pandas as pd
```

```python
docs = [
    "Data Science is an interdisciplinary field.",
    "Machine LEarning is a subset of data science.",
    "Data science uses machien learning algorithms.",
    "Statistics is an important aspect of data science."
]

#Creating a CountVectorizer object to count word frequencies.
count_vectorizer = CountVectorizer()

#Generating the Ter-Document Matrix (TDM) / Document-Term Matrix (DTM)
#The matrix contains the frequency of each word in each document
#Here, the TDM and DTM are the same; the naming depends on the context
term_documents_matrix = count_vectorizer.fit_transform(docs)

#Converting the matix to a DataFrame for better visualization
tdm_df = pd.DataFrame(term_documents_matrix.toarray(),
                      columns=count_vectorizer.get_feature_names_out())

#Print the TDM
print("Term Document Matrix (TDM) / Document-Term Matrix (DTM) :")
print(tdm_df)
print("\n")

#Explanation:
#Rows represent documents and columns represent (words).
#Each cell (i,j) contains the frequency of the word j in document i.

print("****************************************************************************")

#Applying TF-IDF transformation
tfidf_transformer = TfidfTransformer()
tfidf_matrix = tfidf_transformer.fit_transform(term_documents_matrix)

#Convert the matrix to a dataframe 
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(),columns = count_vectorizer.get_feature_names_out())

#Print the tfidf_df
print("TF-IDF :")
print(tfidf_df)
print("\n")

#Explanation:
#TF-IDF (term frequency INverse document frequency) normalizes the word frequencies.
#by accounting for the frequency of the word across all documents.
#This helps in reducing the impact of common words and highlighting important words.
```
