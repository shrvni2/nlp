```python
import spacy
from nltk.corpus import wordnet

nlp = spacy.load("en_core_web_md")

def word_similarity(word1,word2):
    """
    Calculate similarity between two words using Wordnet from NLTK
    """
    
    synset1 = wordnet.synsets(word1)[0]
    synset2 = wordnet.synsets(word2)[0]
    similarity = synset1.wup_similarity(synset2)
    
    return similarity


def sentence_similarity(sentence1,sentence2):
    """
    Calculate similarity between two sentences using spacy's pre-trained model.
    """
    
    doc1 = nlp(sentence1)
    doc2 = nlp(sentence2)
    similarity = doc1.similarity(doc2)
    
    return similarity
```

```python
word_similarity('dancing','singing')
```

```python
sentence_similarity('I have a dog','Cat are not good')
```

```python
word1 = 'Cat'
word2 = 'Dog'
print(f"Similarity between {word1} and {word2} : {word_similarity(word1,word2)}")
```

```python
sentence1 = 'The quick brown fox jumps over the lazy dog'
sentence2 = 'A fast dark fox leaps over a sleepy dog'
print(f"Similarity between sentences:\n'{sentence1}'\n and \n'{sentence2}':\n {sentence_similarity(sentence1,sentence2)}' ")
```

```python
word1 = 'Dear'
word2 = 'Deer'
print(f"Similarity between {word1} and {word2} : {word_similarity(word1,word2)}")
```
