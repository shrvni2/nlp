```python
text = [
    "Artificial intelligence is transforming the world. Many companies are investing in AI research and development.",
    "The new iPhone was released last week, and it features a better camera, improved battery life, and faster processing.",
    "Climate change is one of the most pressing issues of our time. Governments worldwide are working to reduce carbon emissions.",
    "The stock market saw significant gains today, with technology stocks leading the way.",
    "Advances in medical technology have led to better treatments for chronic diseases, improving patient outcomes."
]
```

```python
# Intialize spacy 'en' model, keeping only tagger component need for lemmatization
nlp = spacy.load('en_core_web_md')

# Tokenization, removing stopword, non-alphabetic characters and lemmatization
stop_words = set(stopwords.words('english'))
lemmatizater = WordNetLemmatizer()

def preprocess(text):
    text = text.lower() # convert to lower
    tokens = word_tokenize(text) # tokenize
    tokens = [lemmatizater.lemmatize(word) for word in tokens if word not in stop_words] #lemmatize and remove 
    return tokens

processed_docs = [preprocess(doc) for doc in text]
```

```python
# Create Dictonary 
id2word = corpora.Dictionary(processed_docs)

corpus = [id2word.doc2bow(text) for text in processed_docs]

# view the corpus
print(corpus)
```

```python
# build LDA model
lda_model = LdaModel(corpus=corpus,
                     id2word=id2word,
                     num_topics=3,
                     random_state=42,
                     update_every=1,
                     chunksize=10,
                     passes=10,
                     alpha='auto')

# Print the keyword in 2 topics
topics = lda_model.print_topics()
for topic in topics:
    print(topic)
```

```python
# Install Required Libraries
#!pip install gensim nltk

# Import Necessary Libraries
import gensim
from gensim import corpora
from gensim.models import LdaModel
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# # Download required NLTK datasets for tokenization and stopwords
# nltk.download('punkt')
# nltk.download('stopwords')

# Create a Corpus of 10 Documents
# This is a small set of text documents that we want to analyze for topic modeling
documents = [
 "Apple is looking at buying U.K. startup for $1 billion",
 "Apple buys U.K. startup for $1 billion",
 "San Francisco considers banning sidewalk delivery robots",
 "London is a very big city in the United Kingdom",
 "Football is a popular sport in Europe",
 "The price of oil has increased by 10% last year",
 "Technology companies are growing fast",
 "The global economy is expected to grow this year",
 "Apple is a technology company with a focus on innovation",
 "The United Kingdom is known for its rich history and culture"
]

# Preprocess the Text
# This function tokenizes the text, converts it to lowercase, and removes stopwords
stop_words = set(stopwords.words('english'))

def preprocess(doc):
 # Tokenize the document into words
 tokens = word_tokenize(doc.lower())
 # Filter out non-alphabetic words and stopwords
 tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
 return tokens

# Apply preprocessing to all documents in the corpus
processed_docs = [preprocess(doc) for doc in documents]
print(processed_docs)
print("---------------"*5)
# Create a Dictionary and Corpus
# Dictionary maps each word to a unique ID
dictionary = corpora.Dictionary(processed_docs)
print(dictionary)
print("---------------"*5)

# Corpus is a list of Bag-of-Words (BoW) representations of the documents
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
print(corpus)
print("---------------"*5)

# Perform LDA Topic Modeling
# LDA (Latent Dirichlet Allocation) is used to discover topics within the corpus
num_topics = 3 # Number of topics you want to extract (can be adjusted)
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)

# Print the Topics
# This prints the top words associated with each topic
topics = lda_model.print_topics(num_words=4)
for idx, topic in topics:
 print(f"Topic {idx}: {topic}")

# Assign Each Document to Topics
# For each document, we print the distribution of topics it belongs to
for i, doc in enumerate(corpus):
 topics_in_doc = lda_model.get_document_topics(doc)
 print(f"Document {i+1} is associated with topics: {topics_in_doc}")
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\Shrawani\Downloads\BBC News Train.csv")  # Adjust the file path
print(df.head())
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

nltk.download('stopwords')
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()  # Convert to lowercase
    text = text.split()  # Tokenize
    text = [lemmatizer.lemmatize(word) for word in text if word not in stopwords.words('english')]
    return ' '.join(text)

df['cleaned_text'] = df['Text'].apply(clean_text)
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)
tfidf_matrix = tfidf.fit_transform(df['cleaned_text'])
```

```python
from sklearn.decomposition import LatentDirichletAllocation

# Set the number of topics
n_topics = 5  # Adjust based on experimentation
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(tfidf_matrix)
```

```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

tfidf_feature_names = tfidf.get_feature_names_out()
display_topics(lda, tfidf_feature_names, 10)  # Display top 10 words per topic
```

```python
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.dictionary import Dictionary
from gensim.models.ldamodel import LdaModel

# Convert the dataset into a Gensim-friendly format
tokenized_text = [text.split() for text in df['cleaned_text']]
dictionary = Dictionary(tokenized_text)
corpus = [dictionary.doc2bow(text) for text in tokenized_text]

# Train LDA with Gensim
lda_model = LdaModel(corpus=corpus, num_topics=n_topics, id2word=dictionary, passes=15)

# Compute coherence score
coherence_model = CoherenceModel(model=lda_model, texts=tokenized_text, dictionary=dictionary, coherence='c_v')
coherence_score = coherence_model.get_coherence()
print(f"Coherence Score: {coherence_score}")
```

```python
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

pyLDAvis.enable_notebook()  # Enables inline visualization
vis = gensimvis.prepare(lda_model, corpus, dictionary)
vis
```

```python
def categorize_article(text):
    bow = dictionary.doc2bow(text.split())
    topic_distribution = lda_model.get_document_topics(bow)
    return max(topic_distribution, key=lambda x: x[1])[0]  # Return topic with highest weight

df['predicted_category'] = df['cleaned_text'].apply(categorize_article)
```
